### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]

# Promedio
total = 0
for valor in esperanzas_de_vida:
    total = total + valor

cantidad = 0
for _ in esperanzas_de_vida:
    cantidad = cantidad + 1

promedio = total / cantidad

print(f"Suma total:         {total}")
print(f"Cantidad de países: {cantidad}")
print(f"Promedio:           {promedio}")

# Máximo
maximo = esperanzas_de_vida[0]
for valor in esperanzas_de_vida:
    if valor > maximo:
        maximo = valor

print(f"Valor máximo: {maximo}")

# Mínimo
minimo = esperanzas_de_vida[0]
for valor in esperanzas_de_vida:
    if valor < minimo:
        minimo = valor

print(f"Valor mínimo: {minimo}")
````

**Ejercicio 1.3** — Crea una función llamada `clasificar_esperanza_vida` que reciba un número (años) y retorne:
- `"Alta"` si es >= 75
- `"Media"` si es entre 60 y 74.9
- `"Baja"` si es < 60

Prueba la función con al menos 5 valores distintos e imprime los resultados.

````python
# Tu código aquí
def clasificar_esperanza_vida(años):
    if años >= 75:
        return "Alta"
    elif años >= 60:
        return "Media"
    else:
        return "Baja"

# Pruebas
print(clasificar_esperanza_vida(83.2))
print(clasificar_esperanza_vida(67.8))
print(clasificar_esperanza_vida(49.3))
print()
# 5 valores diferentes
print(clasificar_esperanza_vida(70.2))
print(clasificar_esperanza_vida(36.6))
print(clasificar_esperanza_vida(18.0))
print(clasificar_esperanza_vida(15.3))
print(clasificar_esperanza_vida(0.1))
# la ultima es porque pense en africa 
````

### 2.2 — Limpieza de datos

🔧 **Ejercicio 2.2** — El dataset tiene algunos valores nulos (¡datos reales siempre tienen este problema!).  
1. Rellena los valores nulos de columnas **numéricas** con la **media** de cada columna
2. Verifica que ya no queden nulos
3. Explica en una celda de texto: ¿Por qué usamos la media y no simplemente eliminamos esas filas?

````python
# Se rellena los valores nulos de las columnas numericas con la media de cada columna
columnas_numericas = df.select_dtypes(include="float64").columns
df[columnas_numericas] = df[columnas_numericas].fillna(df[columnas_numericas].mean())

print(df.isnull().sum())
````

**Responde aquí** (doble clic para editar):

*¿Por qué usamos la media para rellenar nulos en vez de eliminar las filas?*

Se usa la media porque borrar filas no es tan buena idea, porque se puede perder información. Por ejemplo, si hay 159 datos y se eliminan 10, se pierde una parte del dataset, y eso puede afectar el análisis, sobre todo si esos datos tenían algo en común.

Por eso, en vez de borrar, se rellenan los valores faltantes con el promedio. No es perfecto, pero ayuda a mantener todos los datos sin cambiar mucho los resultados.

Además, como los datos faltantes son pocos, usar la media es lo más común en estos casos. 

### 2.3 — Estadísticas descriptivas

### 2.5 — Agrupaciones

🔧 **Ejercicio 2.5** — Usa `groupby()` para:
1. Calcular la **esperanza de vida promedio** por `region`
2. Calcular el **gasto en salud promedio** por `nivel_ingresos`
3. Calcular el **número de países** por región
4. ¿Qué región tiene la mayor esperanza de vida promedio? ¿Y la menor?

````python
# Tu código aquí

#1 esperanza de vida promedio
ev_por_region = df.groupby("region")["esperanza_vida"].mean()
print(ev_por_region)

#2 gasto en salud promedio por nivel_ingresos
gasto_por_ingresos = df.groupby("nivel_ingresos")["gasto_salud_usd"].mean()
print(gasto_por_ingresos)

#3 número de paises por región
paises_por_region = df.groupby("region")["pais"].count()
print(paises_por_region)

#4 región con mayor y menor esperanza de vida promedio
region_mayor = ev_por_region.idxmax()
region_menor = ev_por_region.idxmin()

print(f"Mayor esperanza de vida: {region_mayor} ({ev_por_region[region_mayor]:.2f} años)")
print(f"Menor esperanza de vida: {region_menor} ({ev_por_region[region_menor]:.2f} años)")

````
